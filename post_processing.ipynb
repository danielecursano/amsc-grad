```python
points = pd.read_csv("output.csv")
```

```python
plt.figure(figsize=(6,5))
scatter = plt.scatter(points["x"], points["y"], c=points["u"], cmap='viridis', s=30)  # s = marker size
plt.colorbar(scatter, label='u(x,y)')  # Add colorbar
plt.xlabel('x')
plt.ylabel('y')
plt.title('u(x,y)')
plt.axis('equal')  # keep aspect ratio
plt.show()
```

```python
plt.figure(figsize=(6,5))
scatter = plt.scatter(points["x"], points["y"], c=points["true_u"], cmap='viridis', s=30)  # s = marker size
plt.colorbar(scatter, label='u(x,y)')  # Add colorbar
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exact solution')
plt.axis('equal')  # keep aspect ratio
plt.show()
```

```python
points["abs_err"] = abs(points["u"]-points["true_u"])
```

```python
plt.figure(figsize=(6,5))
scatter = plt.scatter(points["x"], points["y"], c=points["abs_err"], cmap='viridis', s=30)  # s = marker size
plt.colorbar(scatter, label='abs err')  # Add colorbar
plt.xlabel('x')
plt.ylabel('y')
plt.title('Absolute error')
plt.axis('equal')  # keep aspect ratio
plt.show()
```
